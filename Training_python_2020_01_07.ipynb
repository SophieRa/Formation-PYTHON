```python
import matplotlib.pyplot as plt

fig = plt.figure() #create the figure

ax = fig.add_subplot(1,1,1)  #create a graph

# Create bars : y values
bars1 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==1,'USUBJID']         #create ARMN 1 bars
bars2 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==2,'USUBJID']         #create ARMN 2 bars
bars3 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==3,'USUBJID']         #create ARMN 3 bars

 
# The X position of bars: x values
r1 = [1,4,7,10]         #fix position for 'ARMN 1' bars
r2 = [2,5,8,11]         #fix position for 'ARMN 2' bars
r3 = [3,6,9,12]         #fix position for 'ARMN 2' bars

# Create barplot
ax.bar(r1, bars1, color = 'blue', label='ARM = 1')
ax.bar(r2, bars2, color = 'red', label='ARM = 2')
ax.bar(r3, bars3, color = 'green', label='ARM = 3')

 
# Create legend
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure() #create the figure

ax1 = fig.add_subplot(1,3,1)  #create a graph (row, col, nb of graphes)
ax2 = fig.add_subplot(1,3,2)  #create a graph
ax3 = fig.add_subplot(1,3,3)  #create a graph

# Create bars
barWidth = 0.9
bars1 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==1,'USUBJID']         #create ARMN 1 bars
bars2 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==2,'USUBJID']         #create ARMN 2 bars
bars3 = TMadre_bar_plot.loc[TMadre_bar_plot['ARMN']==3,'USUBJID']         #create ARMN 3 bars

 
# The X position of bars
r1 = [1,4,7,10]         #fix position for ARMN 1 bars
r2 = [2,5,8,11]         #fix position for ARMN 2 bars
r3 = [3,6,9,12]         #fix position for ARMN 2 bars

# Create barplot
ax1.bar(r1, bars1, width = barWidth, color = 'blue', label='ARM = 1')
ax2.bar(r2, bars2, width = barWidth, color = 'red', label='ARM = 2')
ax3.bar(r3, bars3, width = barWidth, color = 'green', label='ARM = 3')

 
# Create legend
ax1.legend()
ax2.legend()
ax3.legend()
plt.show()
```

### detail of data preparation

```python
# prepare the data
max_values = TMadsl.groupby('ARM')[['x'+str(i) for i in range(1,6)]].max()
max_values.loc['Experimental 1',:].values                    # list of values to plot
values=max_values.loc['Experimental 1',:].values.tolist()    # transform data into list
values += values[:1]                                         # to cycle the values : last value = first value

# prepare the plot
max_value = TMadsl[['x'+str(i) for i in range(1,6)]].max().max()
str((max_value).round(2))                                    # max value for the axes
str((max_value/4).round(2))                                 # 1st tick
str((max_value/2).round(2))                                 # 2st tick
str((max_value*3/4).round(2))                                 # 3t tick
```

### plot

```python
# packages import
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
 
# ----------- prepare data

# extract the max values of X1 to x5
# to fix the plot limits and 
max_value = TMadsl[['x'+str(i) for i in range(1,6)]].max().max()

# values to plot: separte by 'ARM'
max_values = TMadsl.groupby('ARM')[['x'+str(i) for i in range(1,6)]].max()

# ------- PART 1: Create background
 
# number and name of the variables
N = 5
categories=['x'+str(i) for i in range(1,6)]

 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([max_value/4,max_value/2,max_value*3/4], [str((max_value/4).round(2)),
                                                     str((max_value/2).round(2)),
                                                     str((max_value*3/4).round(2))], color="grey", size=7)
plt.ylim(0,max_value)
 
 
# ------ PART 2: Add plots

 
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=max_values.loc['Experimental 1',:].values.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Experimental 1')
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=max_values.loc['Experimental 2',:].values.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Experimental 2')
ax.fill(angles, values, 'r', alpha=0.1)
 
    
# Ind3
values=max_values.loc['Placebo',:].values.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Placebo')
ax.fill(angles, values, 'g', alpha=0.1)


# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```

## <font color=blue>Volcano</font> plot

### data generation

```python
import numpy as np
import pandas as pd
x = np.random.normal(loc=1, scale=0.2, size=100)
y = np.power(10, -(6 * np.absolute(x - 1) +  np.random.normal(0, 1*np.absolute(x - 1), 100)))
data = pd.DataFrame(x, y).reset_index()
data.columns=['x','y']
data['annot']=['bm' + str(i+1) for i in range (100)]
# add the -10logP column
data['-10log(p)']=-np.log10(data['x'])
# add the information on regulation 
UP = ((data['y']>=1.25)&(data['x']<=0.05))
DOWN = ((data['y']<=0.75)&(data['x']<=0.05))
data.loc[:,'Regulated ?']= 'NOT'
data.loc[UP,'Regulated ?']= 'UP'
data.loc[DOWN,'Regulated ?']= 'DOWN'
# to get the down then not then up : fixed color position in the graph
data = data.sort_values('Regulated ?')
```

### plot with `seaborn`

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10)) #changing the size of the figure 

sns.set(style="ticks", palette="bright") #changing the style and color of the figure 

ax = sns.scatterplot(x='y', y='-10log(p)', data=data, hue='Regulated ?', palette=['g','black','r'])

plt.legend(loc='upper center')
plt.plot([1.25, 1.25], [0, 3], color = 'grey', linestyle = 'dashed', linewidth=1)
plt.plot([0.75, 0.75], [0, 3], color = 'grey', linestyle = 'dashed', linewidth=1)
plt.plot([0, 2], [-np.log10(0.05),-np.log10(0.05)], color = 'grey', linestyle = 'dashed', linewidth=1)

plt.show()
```

### <font color=blue>Survival function</font> calculation  with `statsmodels`

* https://www.statsmodels.org/stable/generated/statsmodels.duration.survfunc.SurvfuncRight.html?highlight=kaplan%20meier

```python
from statsmodels.duration.survfunc import SurvfuncRight
```

```python
help(SurvfuncRight)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
gb = TM_KP2.groupby("SEX")
ax = plt.axes()
sexes = []
for g in gb:
    sexes.append(g[0])
    sf = SurvfuncRight(g[1]['AVAL'],g[1]['status'])
    sf.plot(ax)
    
li = ax.get_lines()
li[1].set_visible(False)
li[3].set_visible(False)
plt.figlegend((li[0], li[2]), sexes, "center")

plt.ylim(0, 1)
ax.set_ylabel("Proportion alive")
ax.set_xlabel("Days since enrollment")

plt.show()
```

##  <font color=blue>PCA</font> on breast_cancer data

### Processing data:  <font color=blue>standardisation</font> with `scikit-learn`

```python
from sklearn.datasets import load_breast_cancer
breast = load_breast_cancer()
breast_data = breast.data
breast_labels = breast.target
labels = np.reshape(breast_labels,(569,1))

# concatenation of data and target (0, 1 according the output)
final_breast_data = np.concatenate([breast_data,labels],axis=1)

# conversion in dataframe
import pandas as pd
breast_dataset = pd.DataFrame(final_breast_data)
features = breast.feature_names

features_labels = np.append(features,'label')
breast_dataset.columns = features_labels

breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)

# standardisation of the data 
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:,features].values

x = StandardScaler().fit_transform(x) # normalizing the features

# verification standardisation
np.mean(x),np.std(x)
```

###  <font color=blue>PCA</font> calculation with `scikit-learn`

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=3)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2','principal component 3'])
# Variance each principal component holds after projecting the data to a lower dimensional subspace.
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

###  <font color=blue>PCA</font>  plot with `matplotlib`

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
plt.show()
```
